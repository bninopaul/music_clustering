## Internal Evaluation

### Inertia
Inertia is a metric that is used to estimate how close the data points in a cluster are. This is calculated as the sum of squared distance for each point to it's closest centroid, i.e., its assigned cluster center. The intution behind inertia is that clusters with lower inertia are better as it means closely related points form a cluster.Inertia is calculated by scikit-learn by default.

```python
print "Inertia for KMeans with 3 clusters = %lf " % kmeans_3.inertia_
print "Inertia for KMeans with 5 clusters =  %lf " % kmeans_5.inertia_
print "Inertia for KMeans with 7 clusters = %lf " % kmeans_7.inertia_
```

### Silhouette Score 
The silhouette score measures how close various clusters created are. A higher silhouette score is better as it means that we dont have too many overlapping clusters. The silhouette score can be computed using [sklearn.metrics.silhouette_score](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html#sklearn.metrics.silhouette_score) from scikit learn.

```python
from sklearn.metrics import silhouette_score

# NOTE: Use 500 sample points to calculate the silhouette score
def get_silhouette_score(data, model):
    return silhouette_score(data, model.labels_, sample_size= 500)

print "Silhouette Score for KMeans with 3 clusters = %lf" % get_silhouette_score(feature_matrix, kmeans_3)
print "Silhouette Score for KMeans with 5 clusters = %lf " % get_silhouette_score(feature_matrix, kmeans_5)
print "Silhouette Score for KMeans with 7 clusters = %lf " % get_silhouette_score(feature_matrix, kmeans_7)
```

## External Evaluation

```python
# Return a data structure that contains artist_id, artist_name, top tag, cluster_label for every artist
def join_data_prediction(normalized_data, kmeans_model):
    df = normalized_data.copy()
    df["cluster_label"] = kmeans_model.labels_
    return df

# Run the function for all the models
kmeans_3_joined = join_data_prediction(normalized_data, kmeans_3)
kmeans_5_joined = join_data_prediction(normalized_data, kmeans_5)
kmeans_7_joined = join_data_prediction(normalized_data, kmeans_7)
```

```python
#We need to assign what genres belong to each of the cluster label.
def assign_cluster_tags(joined_data):
    df =  joined_data.groupby("cluster_label").apply(lambda x: pd.value_counts(x["freq_fin"])[:5])
    clusters = pd.DataFrame([[i,df[i].index.tolist()] for i in joined_data.cluster_label.unique()],
                           columns = ["cluster_label", "top5tags"]).sort("cluster_label").reset_index(drop = True)
    return clusters
    
kmeans_3_genres = assign_cluster_tags(kmeans_3_joined)
kmeans_5_genres = assign_cluster_tags(kmeans_5_joined)
kmeans_7_genres = assign_cluster_tags(kmeans_7_joined)
```

###Purity and Accuracy
Two commonly used metrics used for evaluating clustering using external labels are purity and accuracy. **Purity** measures the frequency of data belonging to the same cluster sharing the same class label i.e. if we have a number of items in a cluster how many of those items have the same label ? Meanwhile, **accuracy** measures the frequency of data from the same class appearing in a single cluster i.e. of all the items which have a particular label what fraction appear in the same cluster ?

```python
def get_cluster_purity(joined_data, cluster_top5tags):
    df = pd.merge(joined_data, cluster_top5tags, how = "left", on = "cluster_label")
    df["in_top5tags"] = df[["freq_fin", "top5tags"]].apply(lambda x: x["freq_fin"] in x["top5tags"], axis = 1)
    cluster_group = df.groupby("cluster_label")
    purity = np.mean(cluster_group.in_top5tags.sum()/cluster_group.in_top5tags.count())
    return purity
print "Purity for KMeans with 3 centers %lf " % get_cluster_purity(kmeans_3_joined, kmeans_3_genres)
print "Purity for KMeans with 5 centers %lf " % get_cluster_purity(kmeans_5_joined, kmeans_5_genres)
print "Purity for KMeans with 7 centers %lf " % get_cluster_purity(kmeans_7_joined, kmeans_7_genres)
```

```python
def get_accuracy(joined_data):
    unique_tags = joined_data.freq_fin.unique()
    correct_cluster = 0
    for tag in unique_tags:
        correct_cluster+=max(joined_data[joined_data["freq_fin"]==tag].groupby("cluster_label").song.count())
    return correct_cluster/len(joined_data)
    
print "Accuracy of KMeans with 3 centers %lf " % get_accuracy(kmeans_3_joined)
print "Accuracy of KMeans with 5 centers %lf " % get_accuracy(kmeans_5_joined)
print "Accuracy of KMeans with 7 centers %lf " % get_accuracy(kmeans_7_joined)
```
