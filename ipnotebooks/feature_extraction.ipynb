## Visualization of the Average of Features 
##### Features are first normalized using numpy linalg.norm and then plot the features for the beginning, middle and end segments.

```python
import re
```

```python
beg_feat = sorted([i for i in whole_dataset.columns if re.search("beg_.*",i)])
mid_feat = sorted([i for i in whole_dataset.columns if re.search("mid_.*",i)])
end_feat = sorted([i for i in whole_dataset.columns if re.search("end_.*",i)])
```

```python
%matplotlib inline
```

```python
features = whole_dataset.columns.tolist()
features.remove("song")
features.remove("artist")
normalized_dataset = pd.concat([whole_dataset[["song", "artist"]],
                                whole_dataset[features]/np.linalg.norm(whole_dataset[features], axis = 0)], axis = 1)
```

```python
plt.plot(np.mean(normalized_dataset[beg_feat], axis = 0), '-o')
plt.plot(np.mean(normalized_dataset[mid_feat], axis = 0), '-o')
plt.plot(np.mean(normalized_dataset[end_feat], axis = 0), '-o')
```

## Getting the Genres of the Artists

```python
from pyechonest import artist
```

```python
artists = sorted(pd.unique(normalized_dataset.artist))
```

```python
import time
genre = []
for i in artists:
    try:
        artst = artist.Artist(i)
        genre.append([i,artst.terms])
        print artists.index(i), i
    except artist.util.EchoNestAPIError:
        time.sleep(100)
        continue
```

```python
genrefilter = filter(lambda x: len(x[1])>0,genre)
artist_genre = pd.DataFrame(genrefilter, columns = ['artist', 'genre'])
```

```python
artist_genre["byfrequency"] = artist_genre.genre.map(lambda x: sorted(map(lambda y: (y['frequency'], y['name']),x), reverse = True),genrefilter)
artist_genre["byweights"] = artist_genre.genre.map(lambda x: sorted(map(lambda y: (y['weight'], y['name']),x), reverse = True),genrefilter)
artist_genre["byfrequencygenres"]  = artist_genre.byfrequency.map(lambda x: map(lambda y: y[1],x))
artist_genre["byweightgenres"]  = artist_genre.byweights.map(lambda x: map(lambda y: y[1],x))
artist_genre["freq_fin"] = artist_genre.byfrequency.map(lambda x: x[0][1])
artist_genre["weights_fin"] = artist_genre.byweights.map(lambda x: x[0][1])
```
